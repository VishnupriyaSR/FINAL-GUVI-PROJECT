```python
def generate_synonyms(word):
    synonyms = []
    for syn in wordnet.synsets(word):
        for lemma in syn.lemmas():
            synonyms.append(lemma.name())
    return synonyms

# Words to find synonyms for
words_to_find_synonyms = ['not satisfied', 'could not happier', 'bad', 'disappointing','avoid','terrible','would not coming back']

# Generate synonyms for each word
synonyms_list = []
for word in words_to_find_synonyms:
    synonyms = generate_synonyms(word)
    synonyms_list.extend(synonyms)

# Select 25 unique synonyms
synonyms_selection = random.sample(sorted(set(synonyms_list)), k=50)
```

```python
def replace_with_synonyms(sentence, words_to_replace, synonyms_selection):
    for word, synonym in zip(words_to_replace, synonyms_selection):
        sentence = sentence.replace(word, synonym)
    return sentence

# Replace specific words with synonyms
for index, row in df.iterrows():
    original_sentence = row['cleaned']
    sentiment= row['Sentiment']
    # Check if the review is negative (rating 1 or 2)
    if sentiment == 0:
        # Replace specific words with synonyms
        new_sentence = replace_with_synonyms(original_sentence, words_to_find_synonyms, synonyms_selection)
        # Update the DataFrame with the new sentence
        df.at[index, 'review_text'] = new_sentence
df.head(30)
```

```python
#1.convert into lower
df_new['review_text']=df_new['review_text'].str.lower()
```

```python
#2.Tokenization-split into tokens
df_new['tokenized_reviews'] = df_new.apply(lambda row: nltk.word_tokenize(row['review_text']), axis=1)
```

```python
def remove_stop_words(words_list):
    return [t for t in words_list if t not in stop_words]
```

```python
df_new['stopwords_removed']=df_new.apply(lambda row:remove_stop_words(row['tokenized_reviews']),axis=1)
```

```python
#4.punctuation removal
import string
punctuations = string.punctuation
def remove_punct(words_list):
    return [t for t in words_list if t not in punctuations]
```

```python
df_new['punct_removed']=df_new.apply(lambda row:remove_punct(row['stopwords_removed']),axis=1)
```

```python
#lemmatization
lm=WordNetLemmatizer()
def lemmatize_list(words_list):
    return [lm.lemmatize(t) for t in words_list]
```

```python
df_new['lemmatized']=df_new.apply(lambda row:lemmatize_list(row['punct_removed']),axis=1)
```

```python
#Convert list to string - lemmatized column
def convert_to_string(words_list):
    return " ".join(words_list)
```

```python
df_new['cleaned']=df_new.apply(lambda row:convert_to_string(row['lemmatized']),axis=1)
```
